```python
import pandas as pd
import numpy as np
import scipy.optimize
import matplotlib.pyplot as plt
%matplotlib inline

schmal = pd.read_csv('data/SR-EELS_Profil_schmal.csv')
breit = pd.read_csv('data/SR-EELS_Profil_breit.csv')
schmal_low = pd.read_csv('data/SR-EELS_Profil_schmal_low.csv')
breit_low = pd.read_csv('data/SR-EELS_Profil_breit_low.csv')

schmal.norm = schmal.Y / schmal.Y.max()
breit.norm = breit.Y / breit.Y.max()
schmal_low.norm = schmal_low.Y / schmal_low.Y.max()
breit_low.norm = breit_low.Y / breit_low.Y.max()
```

```python
plt.figure(figsize=(6,4))
plt.plot(breit.X, breit.Y, color='r', label='SpecMag = 315, QSinK7 = -12\%')
plt.plot(schmal.X, schmal.Y, color='b', label='SpecMag = 200, QSinK7 = \ \enspace0\%')
plt.plot(breit_low.X, breit_low.Y, 'r', label='')
plt.plot(schmal_low.X, schmal_low.Y, 'b', label='')
plt.xlim(0, 4096)
plt.xticks(range(0, 4097, 512))
plt.ylim(0, 1.3 * schmal.Y.max())
plt.xlabel('Position auf der Kamera [px]')
plt.ylabel('Intensität [counts]')
plt.legend(loc=2)
plt.savefig('Bilder/SR-EELS_Grenzwert')
plt.show()
```
