```python
import pandas as pd
import numpy as np
import scipy.optimize
import matplotlib.pyplot as plt
%matplotlib inline

df_194 = pd.read_csv('data/EELS-Cal-SpecMag125/EELS #C# 125x 194eV I5 B1 3x3s.msa', names=('X', 'Y'), comment='#')
df_284 = pd.read_csv('data/EELS-Cal-SpecMag125/EELS #C# 125x 284eV I5 B1 3x3s.msa', names=('X', 'Y'), comment='#')
df_374 = pd.read_csv('data/EELS-Cal-SpecMag125/EELS #C# 125x 374eV I5 B1 3x3s.msa', names=('X', 'Y'), comment='#')
```

```python
plt.figure(figsize=(10,5))
plt.plot(df_194.index, df_194.Y, label='$194\,\mathrm{eV}$', c='r')
plt.plot(df_284.index, df_284.Y, label='$284\,\mathrm{eV}$', c='g')
plt.plot(df_374.index, df_374.Y, label='$374\,\mathrm{eV}$', c='b')
plt.xlim(0, 4096)
plt.xticks(range(0, 4097, 512))
plt.xlabel('Position auf der Kamera [px]')
plt.yscale('log')
plt.ylabel('Intensität [arb. u.]')
plt.legend(loc='best')
plt.savefig('Bilder/EELS-Cal_Kohlenstoff.pdf')
plt.show()
```
