```python
import pandas as pd
import numpy as np
import scipy.optimize
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline

df = pd.read_csv('data/100Âµm_pos_demag0,25.csv', sep=',')
```

```python
lin_func = lambda y, m, b: m * y + b
popt, pcov = scipy.optimize.curve_fit(lin_func, df.y_pos, df.x_pos, p0=(1, 0))
print('Geradengleichung: f(y) = %.2f * y + %.0f' % (*popt,))
y_fit = np.linspace(0, 4096, 1000)
x_fit = [lin_func(y, *popt) for y in y_fit]
```

```python
img = mpimg.imread('data/Spalt_demag0,25.png')

f = plt.figure(figsize=(12, 6))
plt.xlabel('x-Position auf Kamera [px]')

ax1 = plt.subplot(121)
plt.xlabel('x-Position auf Kamera [px]')
plt.ylabel('y-Position auf Kamera [px]')
plt.xticks(range(0, 4097, 512))
plt.yticks(range(0, 4097, 512))
plt.imshow(img, cmap='gray', aspect=1)
plt.scatter(df.x_pos, df.y_pos, marker='x', c='lightgreen')
plt.plot(x_fit, y_fit, c='red')
plt.xlim(0, 4096)
plt.ylim(4096, 0)
plt.grid(alpha=0.33)

ax2 = plt.subplot(122)
plt.xlabel('x-Position auf Kamera [px]')
f.subplots_adjust(hspace=0)
plt.setp([a.get_yticklabels() for a in f.axes[1:]], visible=False)
plt.xticks(range(0, 4097, 512))
plt.yticks(range(0, 4097, 512))
plt.imshow(img, cmap='gray', aspect=0.25)
plt.scatter(df.x_pos, df.y_pos, marker='x', c='lightgreen')
plt.plot(x_fit, y_fit, c='red')
plt.xlim(0, 4096)
plt.ylim(4096, 0)
plt.grid(alpha=0.33)
plt.xlim(1536, 2560)
plt.savefig('Bilder/SR-EELS_Position_Rundblende.pdf')
plt.show()
```
