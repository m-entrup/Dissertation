```python
import pandas as pd
import numpy as np
import scipy.optimize
import matplotlib.pyplot as plt
%matplotlib inline

df_crfe_dünn = pd.read_csv('data/ImgSpec_low/CrFe_dünn.csv')
df_cr_dünn = pd.read_csv('data/ImgSpec_low/Cr_dünn.csv')
df_fe_dünn = pd.read_csv('data/ImgSpec_low/Fe_dünn.csv')
df_si_dick = pd.read_csv('data/ImgSpec_low/Si_dick.csv')
df_si_dick.Yn = df_si_dick.Y0 / df_si_dick.Y0.max() * df_crfe_dünn.Y0.max()
df_cr_oplasmon = pd.read_csv('data/ImgSpec_low/Cr_Oberflächenplasmon.csv')
df_cr_oplasmon.Yn = df_cr_oplasmon.Y0 / df_cr_oplasmon.Y0[18] * df_crfe_dünn.Y0[18]
```

```python
plt.figure(figsize=(5,5))
plt.bar(df_crfe_dünn.X0, df_crfe_dünn.Y0, width=2, alpha=0.25, color='k', label='')
plt.scatter(df_crfe_dünn.X0, df_crfe_dünn.Y0, alpha=0.5, color='k', label='Chrom und Eisen (dünn)')
plt.scatter(df_cr_dünn.X0, df_cr_dünn.Y0, alpha=0.5, color='g', label='Chrom-Schicht (dünn)')
plt.scatter(df_fe_dünn.X0, df_fe_dünn.Y0, alpha=0.5, color='r', label='Eisen-Schicht (dünn)')
plt.xlim(4, 70)
plt.xlabel('Energieverlust [eV]')
plt.ylabel('Intensität [arb. u.]')
plt.legend(loc='best')
plt.savefig('Bilder/ImgSpec_low_Spectrum.pdf')
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.bar(df_crfe_dünn.X0, df_crfe_dünn.Y0, width=2, alpha=0.25, color='k', label='')
plt.bar(df_si_dick.X0, df_si_dick.Yn, alpha=0.25, width=2, color='b', label='')
plt.scatter(df_crfe_dünn.X0, df_crfe_dünn.Y0, alpha=0.5, color='k', label='Chrom und Eisen (dünn)')
plt.scatter(df_si_dick.X0, df_si_dick.Yn, alpha=0.5, color='b', label='Silizium (dick)')
plt.xlim(4, 46)
plt.ylim(0, 4250)
plt.xlabel('Energieverlust [eV]')
plt.ylabel('Intensität [arb. u.]')
plt.legend(loc='upper center')
plt.savefig('Bilder/ImgSpec_low_Spectrum2.pdf')
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.bar(df_cr_dünn.X0, df_cr_dünn.Y0, width=2, alpha=0.25, color='k', label='')
plt.bar(df_cr_oplasmon.X0, df_cr_oplasmon.Yn, width=2, alpha=0.25, color='g', label='')
plt.scatter(df_cr_dünn.X0, df_cr_dünn.Y0, alpha=0.5, color='k', label='Chrom (hell)')
plt.scatter(df_cr_oplasmon.X0, df_cr_oplasmon.Yn, alpha=0.5, color='g', label='Chrom (dunkel)')
plt.text(25, 4500, '$25\, \mathrm{eV}$')
plt.text(17, 4500, '$17\, \mathrm{eV}$')
plt.xlim(4, 70)
plt.xlabel('Energieverlust [eV]')
plt.ylim(0, 4750)
plt.ylabel('Intensität [arb. u.]')
plt.legend(loc='best')
plt.savefig('Bilder/ImgSpec_low_Oberflaechenplasmon.pdf')
plt.show()
```
