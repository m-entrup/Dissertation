```python
import pandas as pd
import numpy as np
import scipy.optimize
import matplotlib.pyplot as plt
%matplotlib inline

df_0 = pd.read_csv('data/SR-EELS_energy-cal/EELS_635eV_0%.csv')
df_30 = pd.read_csv('data/SR-EELS_energy-cal/EELS_635eV_-30%.csv')
df_30['Y_scaled'] = df_30.Y * df_0.Y.max() / df_30.Y.max()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df_0.X, df_0.Y, label='QSinK7 = 0%', c='g')
plt.plot(df_30.X, df_30.Y_scaled, label='QSinK7 = -30%', c='r')
plt.xlim(525, 750)
plt.xlabel('Energieverlust [eV]')
plt.ylim(12e4, 38e4)
plt.ylabel('Intensität [a.u.]')
plt.text(575, 36e4, 'L$_3$-Kante', fontdict=dict(color='g', size=14))
plt.text(584, 34e4, 'L$_2$-Kante', fontdict=dict(color='g', size=14))
plt.text(708, 20e4, 'L$_3$-Kante', fontdict=dict(color='g', size=14))
plt.text(721, 17e4, 'L$_2$-Kante', fontdict=dict(color='g', size=14))
plt.legend(loc='best')
plt.savefig('Bilder/SR-EELS_Energy-Cal_falsch.pdf')
plt.show()
```
